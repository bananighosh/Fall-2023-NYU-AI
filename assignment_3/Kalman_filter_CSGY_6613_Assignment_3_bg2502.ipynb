```python
# Replace 'your_video_url' with the actual URL of the video you want to extract images from
# video_url = 'https://www.youtube.com/watch?v=DhmZ6W1UAv4&ab'

video_url = 'https://www.youtube.com/watch?v=DhmZ6W1UAv4&ab'

# Create a YouTube object
yt = YouTube(video_url, use_oauth=True, allow_oauth_cache=True)

# Get the video stream with the highest resolution
video_stream = yt.streams.get_highest_resolution()

# Get the video title for naming the images
video_title = yt.title

# Create a VideoCapture object
cap = cv2.VideoCapture(video_stream.url)

# Set the frame rate and initialize variables
fps = int(cap.get(cv2.CAP_PROP_FPS))
interval = 10  # Extract a frame every 10 seconds
frame_number = 1

download_path = "./drone_videos/video_2/"
pathlib.Path(download_path).mkdir(parents=True, exist_ok=True)

while True:
    # Read the next frame
    ret, frame = cap.read()

    # Break the loop if the video is finished
    if not ret:
        break

    # Extract a frame every 'interval' seconds
    if frame_number % (fps * interval) == 0:
        # Save the frame as an image
        image_path = f"{download_path}{video_title}_frame_{frame_number}.jpg"
        cv2.imwrite(image_path, frame)
        print(f"Frame {frame_number} saved as {image_path}")

    frame_number += 1

# Release the VideoCapture object
cap.release()
```

```python
## For more than these 2 videos, we will traverse the videos directory one by one and execute the same code

VIDEOS_DIR = os.path.join('.', ROOT_DIR, 'drone_videos', 'videos')
DETECTIONS_DIR = os.path.join('.', 'detections')  # New directory for storing detection frames

video_path = os.path.join(VIDEOS_DIR, 'Drone Tracking 1.mp4')
video_path_out = '{}_out.mp4'.format(video_path)

video_path = os.path.abspath(video_path)
print(video_path)

print(video_path_out)

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()

print(cap.read())

H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

model_path = os.path.join('.', 'runs', 'detect', 'train', 'weights', 'last.pt')

# Load a model
model = YOLO(model_path)  # load a custom model

threshold = 0.5

# Create detections directory if it does not exist
if not os.path.exists(DETECTIONS_DIR):
    os.makedirs(DETECTIONS_DIR)

while ret:
    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    # Save frames with bounding boxes to the detections directory
    frame_filename = os.path.join(DETECTIONS_DIR, 'frame_{}.jpg'.format(ret))  # Unique filename for each frame
    cv2.imwrite(frame_filename, frame)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```

#### For more than one video processing from the videos directory

```python
from google.colab.patches import cv2_imshow

def process_video(video_path):
    cap = cv2.VideoCapture(video_path)
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        cv2_imshow('Video Frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()

def process_videos_in_directory(directory):
    # Get a list of all files in the directory
    file_list = [f for f in os.listdir(directory) if f.endswith('.mp4') or f.endswith('.avi')]

    # Process each video in the directory
    for video_file in file_list:
        video_path = os.path.join(directory, video_file)
        print(f"Processing video: {video_path}")
        process_video(video_path)


# Specify the directory containing the videos
VIDEOS_DIR = os.path.join('.', 'drone_videos', 'videos')

# Process videos in the specified directory
process_videos_in_directory(VIDEOS_DIR)
```

```python
from filterpy.kalman import KalmanFilter
import numpy as np
import cv2
import math

# Initialize Kalman filter parameters
kf = KalmanFilter(dim_x=4, dim_z=2)
kf.F = np.array([[1, 0, 1, 0],
                 [0, 1, 0, 1],
                 [0, 0, 1, 0],
                 [0, 0, 0, 1]])
kf.H = np.array([[1, 0, 0, 0],
                 [0, 1, 0, 0]])
kf.R = np.diag([0.1, 0.1])  # Measurement noise covariance
kf.Q = np.diag([0.01, 0.01, 0.01, 0.01])  # Process noise covariance
kf.P *= 10  # Initial state covariance
kf.x = np.array([0, 0, 0, 0])  # Initial state

count = 0
history_length = 20
predicted_positions = []

def draw_boxes(frame, boxes, identities, categories):
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        identity = int(identities[i])
        category = int(categories[i])

        # Draw bounding box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Draw text (you can customize this based on your needs)
        text = f"ID: {identity}, Category: {category}"
        cv2.putText(frame, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)


VIDEOS_DIR = os.path.join('.', ROOT_DIR, 'drone_videos', 'videos')
DETECTIONS_DIR = os.path.join('.', 'detections')  # New directory for storing detection frames

video_path = os.path.join(VIDEOS_DIR, 'Drone Tracking 1.mp4')
video_path_out = '{}_out.mp4'.format(video_path)

video_path = os.path.abspath(video_path)
print(video_path)

print(video_path_out)
cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'mp4v'), 30.0, (frame_width, frame_height))

while True:
    ret, frame = cap.read()
    count += 1
    if ret:
        detections = np.empty((0, 6))
        result = list(model.predict(frame, conf=0.40))[0]
        bbox_xyxys = result.prediction.bboxes_xyxy.tolist()
        confidences = result.prediction.confidence
        labels = result.prediction.labels.tolist()

        # Update Kalman filter with measurements
        tracker_dets = np.array([kf.update(det[:4]) for det in detections])

        if len(tracker_dets) > 0:
            bbox_xyxy = tracker_dets[:, :4]
            identities = np.arange(len(tracker_dets))  # Assign unique IDs
            categories = tracker_dets[:, 4]

            # Draw bounding boxes and predicted positions
            draw_boxes(frame, bbox_xyxy, identities, categories)

            # Draw predicted positions
            for position in predicted_positions:
                cv2.circle(frame, (int(position[0]), int(position[1])), 5, (0, 0, 255), -1)

            # Keep track of last N predicted positions
            if len(predicted_positions) >= history_length:
                predicted_positions.pop(0)
            predicted_positions.append(tracker_dets[0, :2])

        out.write(frame)
        cv2.imshow('Frame', frame)

        if cv2.waitKey(30) & 0xFF == 27:  # Press 'Esc' to exit
            break

cap.release()
out.release()
cv2.destroyAllWindows()
```
